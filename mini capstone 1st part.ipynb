```python
# Q7: Please find what is the minimum total charges for Los Angeles city.
#     Please find which zip code has the maximum total charges.
```

```python
#Ans7.(a)
#converting object type data into numerical float data type.
raw_df['Total Charges'] = pd.to_numeric(raw_df['Total Charges'],errors='coerce')
```

```python
# selecting only los angeles entry and sorting it in ascending order and extracting required row entry for 1st place.
los_angeles= raw_df[raw_df['City'] == 'Los Angeles']
los_angeles.sort_values('Total Charges')[['City','Total Charges','CustomerID']].iloc[0]
```

```python
#Ans7.(b)
```

```python
#For single entry max total charges is:
max_total_charges = raw_df.sort_values('Total Charges', ascending=False)[['City','Zip Code','Total Charges']].iloc[0]
max_total_charges
```

```python
#For cobine entry max total charges is:
max_total_zip_charges_combine = raw_df.groupby('Zip Code').sum()['Total Charges']
max_total_zip_charges_combine.sort_values(ascending=False)
```

```python
max_total_zip_charges_combine_heighest = max_total_zip_charges_combine.sort_values(ascending=False).iloc[0]
max_total_zip_charges_combine_heighest
```
